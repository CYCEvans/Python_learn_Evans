```python
html = "http://www.coolpc.com.tw/evaluate.php"
res= requests.get(html)
```

```python
res
```

```python
soup = BeautifulSoup(res.text,"html5lib")
```

```python
import re 
```

```python
txt = soup.select("#tbdy > tr")[6].select("optgroup")[0]
```

```python
textList = [textString.get_text() for textString in txt.select("option ")]
```

```python
result = []
for text in textList:
    ssd = dict()
    text.strip()
    ssd["brand"] = text.split(" ")[0]
    ssd["name"] = text.split('/')[0]
    warranty = re.findall("【(\w)年", text)
    ssd["warranty"] = warranty[0] if warranty else ""
    flash = re.findall("\wLC",text)
    ssd["flash"] = flash[0] if flash else ""
    readWay = re.findall("讀:(\d+)M",text)
    ssd["readWay"] = int(readWay[0]) if readWay else 0
    writeWay = re.findall("寫:(\d+)M",text)
    ssd["writeWay"] = int(writeWay[0]) if writeWay else 0    
    cap = re.findall("(\d+)(G|TB)",text)[0]
    
    if cap[1] == "TB":
        size = int(cap[0]) * 1000
    else:
        size = int(cap[0])
    ssd['cap'] = size
    ssd['price'] = int(min(re.findall('\$(\d+)', text)))
    result.append(ssd)
```

```python
result
```

```python
[ssd  for ssd in result if ssd["price"] < 2500]
```
